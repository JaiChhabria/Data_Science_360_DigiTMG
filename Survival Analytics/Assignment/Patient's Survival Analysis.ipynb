Lets load necessary libraries and our dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
plt.rcParams['figure.figsize'] = (12,12)
plt.style.use('ggplot')
```

```python
data = pd.read_csv('Patient.csv')
data.head()
```

```python
data.shape
```

```python
print(data['Eventtype'].value_counts()/data.shape[0])
(data['Eventtype'].value_counts()/data.shape[0]).plot(kind='pie')
```

Follow Up is referring to Time

```python
T = data.Followup
```

```python
# Importing KaplanMeierFitter model to fit the survival analysis
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
```

```python
# Fitting the model on T and Event of the data

kmf.fit(T, event_observed=data.Eventtype)

# plotting the Time Line Estimation plot
kmf.plot()
```
