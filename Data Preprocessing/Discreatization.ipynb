# As we can see apart from `Species` column all other columns are continous columns, we will use following methods to discreatization our columns

We will discreate following columns `Sepal.Length`, `Sepal.Width`, `Petal.Length`, `Petal.Width`

We will use following techniques:
1 - Binarization
2 - Rounding
3 - Binning

## Step 1 - Binarization

```python
# We will use Python qcut, to categorize our value to be either Below Median or Above Median

# We will use this logic for each column and create seperate columns which will show Binarization in our dataset
```

```python
bins_label = ['Below_Median', 'Above_Median'] # Creating Labels for our Binarization
```

Transforming our columns in Dataset

```python
df['Sepal_Length_Binarization'] = pd.qcut(df['Sepal.Length'], q=2,
                                         labels=bins_label) # We are cutting our data into 50% qunatile and we are categorzing our Sepal length to be either above Median or Below median
df
```

```python
df['Sepal_Width_Binarization'] = pd.qcut(df['Sepal.Width'], q=2,
                                        labels=bins_label)
df
```

```python
df['Petal_Length_Binarization'] = pd.qcut(df['Petal.Length'], q=2,
                                         labels=bins_label)
df
```

```python
df['Petal_Width_Binarization'] =pd.qcut(df['Petal.Width'], q=2,
                                       labels=bins_label)
df
```

We have sucessfully Binarized our continous data to be in two categories which is Above Median or Below Median

# Step 2 - Rounding

```python
## We will round the numbers to their nearest Integer
```

```python
## We will load the data again
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
columns_to_drop = ['Unnamed: 0']
df = df.drop(columns=columns_to_drop)
df
```

```python
df['Sepal_Length_Rounded'] = np.array(np.round((df['Sepal.Length'])), dtype='int') # Rounding Sepal.Length column to its nearest integer and saving that information to a new column
df
# We will replicate the same process for other columns in the dataframe
```

```python
df['Sepal_width_Rounded'] = np.array(np.round((df['Sepal.Width'])), dtype='int')
df['Petal_Length_Rounded'] = np.array(np.round((df['Petal.Length'])),dtype='int')
df['Petal_Width_Rounded'] = np.array(np.round((df['Petal.Width'])), dtype='int')
df
```

We have sucessfully Rounded our continous data to its nearest `Integer`

```python
# Lets load the data again
df = pd.read_csv('iris.csv')
df
columns_to_drop = ['Unnamed: 0']
df = df.drop(columns=columns_to_drop)
df
```

```python
df.describe()
```

```python
# from the data set, we can see max value is 7.9 and min value is 0.100000

# We will create an array of 5 values which will be between 0.1 to 8

custom_bin = np.linspace(0.1, 8, 5)
custom_bin
```

```python
# We will provide our custom bin value to our cut to bin values only between these values
```

```python
df['Sepal.Length_Bin'] = pd.cut(df['Sepal.Length'],custom_bin)
df['Speal.Width_Bin'] = pd.cut(df['Sepal.Width'],custom_bin)
df['Petal.Lenght_Bin'] = pd.cut(df['Petal.Length'],custom_bin)
df['Petal.Width_Bin'] = pd.cut(df['Petal.Width'],custom_bin)
df
```
